```python
data.dropna(inplace=True)
# Remove rows with missing values
data.drop_duplicates(inplace=True)
# Remove duplicate rows
```

```python
data.head()
```

```python
from matplotlib import pyplot as plt

# Create a Property_Age column (assuming current year is 2024)
data["Property_Age"] = 2024 - data["Built Year"]

# Plot the average price by property age
data.groupby("Property_Age")["Price"].mean().sort_values(ascending=False).plot(kind='bar')
plt.title("Average Price by Property Age")
plt.xlabel("Property Age (Years)")
plt.ylabel("Average Price")
plt.show()
```

```python
x = data[["number of bedrooms", "number of bathrooms", "living area", "lot area", "Property_Age"]]
y = data["Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    'criterion': ['mse', 'friedman_mse', 'mae'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python


from matplotlib.pyplot import grid
grid.tree=GridSearchCV(estimator=tree_model, param_grid=param_grid)
```

```python
# No need to import grid from matplotlib.pyplot

# Fit the GridSearchCV object to the training data
grid.tree.fit(x_train, y_train)
```

```python
# Display the best parameters found by GridSearchCV
grid.tree.best_params_
```

```python
tree_pred = grid.tree.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
# tree_pred = grid.tree.predict(x_test)
mean_absolute_error(y_test, tree_pred)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predslr = lr.predict(x_test)
```

```python
mean_absolute_error(y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth": [5, 10, 15],
    "n_estimators": [2,3,4,5,6,7,8,9,10],

}
```

```python
grid_rfr = GridSearchCV(rf,param_gridrfr)
```

```python
grid_rfr.fit(x_train, y_train)
# Display the best parameters found by GridSearchCV for Decision Tree
# Display the best parameters found by GridSearchCV for Random Forest
```

```python
grid_rfr.best_params_
```

```python
rfrprediction = grid_rfr.predict(x_test)
```

```python
mean_absolute_error(y_test, rfrprediction)
```
